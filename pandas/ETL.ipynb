### Data injection using pandas and SQLalchemy.

```python
# Define database credentials
# format : postgresql://username:password@localhost:5432/database_name
DB_URL = "postgresql+psycopg2://postgres:1111@localhost:5432/blinkit_customer_analysis_db"

# connection engine.
conn = create_engine(DB_URL)

# file to upload.
files = {
    # 'Data/cleaned/blinkit_orders_cleaned_final.csv' : 'blinkit_orders', # this was injected into the main table in postgres.
    'Data/cleaned/blinkit_customers_cleaned.csv' : 'blinkit_customer_details',
    'Data/cleaned/blinkit_customer_feedback_cleaned.csv' : 'blinkit_customer_feedback',
}

for file_name, table_name in files.items():
    print(f"Uploading {file_name} to table {table_name}...")

    # reading each file with each iteration.
    df=pd.read_csv(file_name)

    # Uploading to postgreSQL database.
    df.to_sql(table_name, conn, if_exists="append", index=False)

    print(f"file {file_name} is successfully uploaded to {table_name}.")

print("All datasets uploaded successfully!")
```
