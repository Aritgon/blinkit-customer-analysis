```python
# changing some capitalized data to lower (string operation)
df['product_name'] = df['product_name'].str.lower()
df['category'] = df['category'].str.lower()
```

```python
df.head()
```

```python
# avg mrp per category.
df.groupby('category')['mrp'].mean().reset_index().round(2)
```

```python
# avg maintaining min stock quantity expense per product category by blinkit. (data of 2023 and 2024)
avg_stock_val = (
    df.groupby('category', as_index=False).agg
    (
        avg_min_stock = ('min_stock_level', 'mean'),
        avg_price = ('price', 'mean')
    )
)

avg_stock_val['avg_maintaining_stock_value'] = (
    avg_stock_val['avg_min_stock'] * avg_stock_val['avg_price']
)

avg_stock_val.sort_values(by = 'avg_maintaining_stock_value', ascending=False).reset_index().round(2)
```

```python
df['profit'] = (df['mrp'] - df['price']).round(2)
```
