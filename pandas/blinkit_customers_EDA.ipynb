```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Soham Gon\Desktop\blinkit analysis\Data\csv\blinkit_customers.csv")

df.head()
```

```python
df['registration_date'] = pd.to_datetime(df['registration_date'], format = '%Y-%m-%d')

df['registration_year'] = df['registration_date'].dt.year

yearly_regt_count = (
    df.groupby('registration_year', as_index=False).agg(
        total_count = ('customer_id', 'count')
    )
)

yearly_regt_count['regt_share'] = (yearly_regt_count['total_count'] / yearly_regt_count['total_count'].sum()) * 100
yearly_regt_count
```
